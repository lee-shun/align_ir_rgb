```python
df_mean_sq = pd.read_csv("/home/ls/m300_rgb_ir_registration/align_ir_rgb/build/mean_sq_error.csv").sort_values('distance')
df_mean_sq
```

```python
df_error = pd.read_csv("/home/ls/m300_rgb_ir_registration/align_ir_rgb/build/error.csv").sort_values('distance')
df_error
```

```python
distance_mean_sq = df_mean_sq['distance']
error_mean_sq = df_mean_sq['mean_sq']
distance_all = df_error['distance']
x_error_all = df_error['x']
y_error_all = df_error['y']
```

```python

font1 = {'family' : 'Times New Roman',
'weight' : 'bold',
'size'   : 15,
}
font2 = {'family' : 'Times New Roman',
'weight' : 'bold',
'size'   : 12,
}


fig = plt.figure(figsize=[10,6])
ax = fig.gca()
x_major_locator=MultipleLocator(1)
y_major_locator=MultipleLocator(1)
ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)
plt.xlim(6,33)

plt.plot(distance_mean_sq, error_mean_sq, c="blue", label='Mean square error', marker='o')
plt.scatter(distance_all, x_error_all, c="red", label='X error', marker='^')
plt.scatter(distance_all, y_error_all, c="green", label='Y error', marker='s')

plt.xlabel('Depth (m)', font1)  
plt.ylabel('Error (pixel)', font1)  
plt.grid()
plt.legend(prop=font2)
plt.show()
fig.savefig("model_reg_error.pdf")
```
